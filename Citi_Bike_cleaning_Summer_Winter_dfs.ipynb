```python
# dependencies
import pandas as pd
```

```python
# read csv dated June 2017
csv1 = "Output/all_combined.csv"
winter_summer = pd.read_csv(csv1)
winter_summer
```

```python
# code source: https://www.geeksforgeeks.org/how-to-drop-one-or-multiple-columns-in-pandas-dataframe/
winter_summer_chopped = winter_summer.drop(['start station id', 
                                            'start station name', 
                                            'end station id', 
                                           'end station name',
                                           'bikeid',
                                           'birth year', 'gender'], axis = 1) 
```

```python
winter_summer_chopped
```

```python
# recalculate the tripduration to minutes
winter_summer_chopped['tripduration_mins'] = winter_summer_chopped['tripduration'] / 60
winter_summer_chopped
```

```python
# output to csv file for easier reading in Tableau
winter_summer_chopped.to_csv('Output/winter_summer_smaller.csv', index=False)
```
